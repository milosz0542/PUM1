## Zadanie 1

```kotlin
val num: Int = 15
var podzielnosc: Boolean = false
for (i in 1 until num+1) {
    podzielnosc = false
    if (i % 3 == 0) {
        print("trzy")
        podzielnosc = true
    }
    if (i % 5 == 0) {
        print("piec")
        podzielnosc = true
    }
    if (!podzielnosc) {
        print(i)
    }

    print("\n")
}
```

## Zadanie 2

```kotlin
fun isPalindrome(x: String): Boolean {
    if (x == x.reversed()) return true
    return false
}

var word: String = "abc"
println(isPalindrome(word).toString())

word = "chujjuhc"
println(isPalindrome(word).toString())
```

```kotlin
fun printPascal(x: Int) {
    for (i in 0 until x) {
        print(" ".repeat(x-i))

        var number = 1
        for (j in 0..i) {
            print("$number ")

            number = number * (i - j) / (j + 1)
        }
        println()
    }
}

printPascal(5)
```

## Zadanie 4

```kotlin
fun isPerfect(x: Int): String {
    var sum_of_dividers: Int = 0

    for (i in 1..x/2) {
        if (x % i == 0) {
            sum_of_dividers += i
        }
    }

    if (sum_of_dividers == x) {
        return "Liczba doskonala"
    } else if (sum_of_dividers > x) {
        return "Liczba obfita"
    } else if (sum_of_dividers < x) {
        return "Liczba niedomiarowa"
    }
    return "Coś nie działa"
}

println(isPerfect(6))
println(isPerfect(12))
println(isPerfect(8))
```

## Zadanie 5

```kotlin
fun isArmstrongNumber(x: Int): Boolean {
    val s = x.toString()
    val n = s.length

    var sum: Int = 0

    for (char in s) {
        val digit = char.digitToInt()

        sum += digit.toDouble().pow(n).toInt()
    }

    if (sum == x) {
        return true
    }

    return false
}

println(isArmstrongNumber(16))
println(isArmstrongNumber(32))
println(isArmstrongNumber(153))
println(isArmstrongNumber(155))
println(isArmstrongNumber(1634))
println(isArmstrongNumber(5000))
```

## Zadanie 6

```kotlin
fun isPrime(x: Int): Boolean {
    if (x < 2) return false

    for (i in 2..x/2) {
        if (x % i == 0) {
            return false
        }
    }

    return true
}

println(isPrime(1))
println(isPrime(2))
println(isPrime(3))
println(isPrime(5))
println(isPrime(6))
```

## Zadanie 7

```kotlin
fun sumEven(x: Int): Int {
    var sum: Int = 0

    for (i in 1..x) {
        if (i % 2 == 0) {
            sum += i
        }
    }

    return sum
}

println(sumEven(2))
println(sumEven(5))
println(sumEven(10))
println(sumEven(11))
```

## Zadanie 8

```kotlin
fun countVowels(str: String): Int {
    val vowels = charArrayOf('a', 'e', 'i', 'o', 'u', 'y')
    var count: Int = 0

    for (i in str.lowercase()) {
        if (i in vowels) {
            count++
        }
    }

    return count
}

println(countVowels("Programowanie"))
println(countVowels("AaBbCcDdEe"))
```
