data class UserInput(val name: String?, val email: String?, val age: String?)
data class UserProfile(
    var name: String = "",
    var email: String = "",
    var age: Int = 0,
    var isAdult: Boolean = false
)

fun buildProfile(input: UserInput?, logs: MutableList<String>): UserProfile? {
    if (input == null) {
        logs.add("Input is null")
        return null
    }

    if (input.name == null) {
        logs.add("Name is null")
        return null
    }
    val name = input.name.trim()
    if (name.length < 3) {
        logs.add("Name too short")
        return null
    }

    if (input.email == null) {
        logs.add("Email is null")
        return null
    }
    val email = input.email.trim().lowercase()
    if (!email.contains("@")) {
        logs.add("Invalid email")
        return null
    }

    if (input.age == null) {
        logs.add("Age is null")
        return null
    }
    val age = input.age.toIntOrNull()
    if (age == null) {
        logs.add("Age is not a number")
        return null
    }

    val profile = UserProfile()
    profile.name = name
    profile.email = email
    profile.age = age
    profile.isAdult = age >= 18

    logs.add("Profile created for $email")
    return profile
}

````kotlin
data class UserInput(val name: String?, val email: String?, val age: String?)
data class UserProfile(
    var name: String = "",
    var email: String = "",
    var age: Int = 0,
    var isAdult: Boolean = false
)

fun buildProfile(input: UserInput?, logs: MutableList<String>): UserProfile? {
    return input?.let {
        val name = it.name?.trim()?.takeIf { n-> n.length >= 3 }
            ?: return null.also { logs.add("Invalid name") }

        val email = it.email?.trim()?.lowercase()?.takeIf { e -> e.contains("@") }
            ?: return null.also { logs.add("Invalid email") }

        val age = it.age?.toIntOrNull()
            ?: return null.also { logs.add("Invalid age") }

        UserProfile().apply {
            this.name = name
            this.email = email
            this.age = age
            this.isAdult = age >= 18
        }.also {
            logs.add("Profile created for $email")
        }
    } ?: run {
        logs.add("Input is null")
        return null
    }
}
var combined_logs: MutableList<String> = mutableListOf<String>()

val JanKowalski: UserProfile? = buildProfile(UserInput("Jan Kowalski", "jan.kowalski@uwr.edu.pl", "25"), combined_logs)
val AdamNowak: UserProfile? = buildProfile(UserInput("Adam Nowak", "adam.nowak@uwr.edu.pl", "35"), combined_logs)

for (i in combined_logs) {
    println(i)
}
````

## Zadanie 3

````kotlin
fun <A> isSorted(lst: List<A>, order: (A, A) -> Boolean): Boolean {
    for(i in 1..lst.size-1) {
        if (!order.invoke(lst[i-1], lst[i])) {
            return false
        }
    }
    return true
}

println(isSorted(listOf(1, 2, 3, 4, 5), {i: Int, j: Int -> i < j}))
println(isSorted(listOf(1, 3, 2, 4, 5), {i: Int, j: Int -> i < j}))
println(isSorted(listOf(1, 2, 3, 4, 5), {i: Int, j: Int -> i > j}))
println(isSorted(listOf(5, 4, 3, 2, 1), {i: Int, j: Int -> i > j}))
````

````kotlin
fun safeParseAndClassify(input: String?): String {
    val result = input?.toIntOrNull() ?: return "BRAK DANYCH"

    if (result % 2 == 0) return "PARZYSTA"
    return "NIEPARZYSTA"
}

println(safeParseAndClassify(null))
println(safeParseAndClassify("123"))
println(safeParseAndClassify("a"))
println(safeParseAndClassify(""))
println(safeParseAndClassify("66"))
````

````kotlin
fun check(N: Int, lst: List<Int>): Int {
    for (i in N until lst.size) {
        val target = lst[i]

        val preamble = lst.subList(i - N, i)

        val seen = hashSetOf<Int>() // Zmienna pomocnicza - Sprawdzamy jakie liczby już "zobaczyliśmy"
        var foundPair = false

        for (num in preamble) {
            val complement = target - num // Obliczamy lst[i] - aktualnie sprawdzaną liczbę

            if (seen.contains(complement)) { // Jeśli ta różnica była już "zauważona" oznacza to, że istnieje opcja na stworzenie liczby lst[i] spośród podzbioru preamble
                foundPair = true
                break
            }
            seen.add(num) // Dodajemy num do hashSet'u seen (czyli ją "zobaczyliśmy")
        }

        if (!foundPair) return target
    }

    return -1
}

/*
    Dlaczego Hashsety?
    Lubię algorytmikę i chciałem w Kotlinie potestować/pobawić się algorytmami.
    Sprawdzanie tego czy daną liczbę da się ułożyć z sumy dwóch liczb z podzbioru L-elementów miałoby złożoność czasową równą O(LxN^2)
    Gdzie L - długość preambuły, N - liczba wszystkich elementów w zbiorze.
    Złożoność czasowa przy skorzystaniu z Hashsetu wynosi O(LxN).
    Wnioski: hashset jest korzystniejszy
 */

println(check(3, listOf(1, 2, 3, 5, 7, 12, 30)))
println(check(2, listOf(1, 2, 3, 5, 7, 12, 30)))
println(check(3, listOf(1, 2, 3, 5, 7, 12)))
````
